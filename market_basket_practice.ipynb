```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import mlxtend
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
data = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/basket_items.csv')
```

```python
data
```

```python
basket = data['items'].str.get_dummies('|')
```

```python
basket
```

```python
baskets = basket.astype(pd.SparseDtype('bool'))
```

```python
baskets
```

```python
frequent_itemsets = apriori(baskets, min_support = 0.1, use_colnames = True)
frequent_itemsets['support_n']=frequent_itemsets['support']*len(baskets)
frequent_itemsets.sort_values('support', ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = .5)
rules
```
